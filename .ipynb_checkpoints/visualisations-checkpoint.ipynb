```python
def outliers(df, replace=np.nan):
    Q1, Q3 = np.percentile(df, [25 ,75])
    IQR = Q3-Q1
    return df.where((df<(Q1- 1.5 * IQR )) | (df > (Q3 + 1.5 * IQR)), replace)

df_new = df.apply(outliers)
df_new.isna().sum()
```

```python
def min_max_values(col):
    top = df[col].idxmax()
    top_obs = pd.DataFrame(df.loc[top])
    
    bottom = df[col].idxmin()
    bot_obs = pd.DataFrame(df.loc[bottom])
    
    min_max_obs = pd.concat([top_obs, bot_obs], axis = 1)
    
    return min_max_obs
```

```python
baseline = {}
for i in df_describe.columns:
    baseline[i] = df_describe[i]
    


baseline_test = {}
for i in df_test_describe.columns:
    baseline_test[i] = df_test_describe[i]
    
print(baseline_test)
```

```python
# compare to baseline
def compare_to_baseline(baseline, test_baseline):
    p_threshold = 0.6
    n_threshold = -0.6
    compare_results = pd.concat([pd.DataFrame.from_dict(baseline), pd.DataFrame.from_dict(test_baseline)], axis = 1)
#     compare_results['value_match'] = np.where(pd.DataFrame.from_dict(baseline) == pd.DataFrame.from_dict(test_baseline), 'True', 'False')
#     if compare_results['value_match'].all(axis=0) and compare_results['value_match'][0] == 'True':
    if pd.DataFrame.from_dict(baseline).equals(pd.DataFrame.from_dict(test_baseline)):
        print('The data is similar to the baseline')
    else:
        compare_results['correlation'] = (compare_results.iloc[:,0]).corr(compare_results.iloc[:,1])
        corr_score = (compare_results.iloc[:,0].corr(compare_results.iloc[:,1]))
        print('Pearsons correlation: %.2f'  %corr_score)  
        if n_threshold < corr_score < p_threshold:
            print('Warning: The statistics of the data do not agree with the baseline')
        
#         corr_results = compare_results.corr()
#         return corr_results
    return compare_results

results = compare_to_baseline(baseline,baseline_test)
results
```

```python
def compare_to_baseline(baseline, test_baseline):
    p_threshold = 0.6
    n_threshold = -0.6
    compare_results = pd.concat([pd.DataFrame.from_dict(baseline, orient = 'index'), pd.DataFrame.from_dict(test_baseline,orient = 'index')], axis = 1)
    if pd.DataFrame.from_dict(baseline).equals(pd.DataFrame.from_dict(test_baseline)):
        print('The data is similar to the baseline')
    else:
        compare_results['correlation'] = (pd.DataFrame.from_dict(baseline, orient = 'index')).corrwith(pd.DataFrame.from_dict(test_baseline,orient = 'index'), axis = 1)
        for index, i in enumerate(compare_results['correlation']):
            if n_threshold <i < p_threshold:
                print(f'Warning: The statistics of the data do not agree with the baseline for {compare_results.T.keys()[index]}')
    return compare_results

results = compare_to_baseline(baseline,baseline_test)
results
```
